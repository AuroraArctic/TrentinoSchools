## Data Preparation

School data are found in vivoscuola website, where it is possible to download the csv version of the data published in OpenDataTrentino. In that website, data was hosted as XML nested file. 

```python
# Import dataset
df = pd.read_csv("../data/trentino/scuole.csv", sep = ";")
```

```python
# Remove Istituti Comprensivi
df = df[~df['Scuola'].isnull()]
```

```python
# Infer school type by the complete name
type = []
for i in range(len(df)):
    s = df.iloc[i, 1].lower()
    t = ""
    if ("asilo" in s) or ("scuola materna" in s):
        t = "Scuola Materna"
    elif "scuola dell'infanzia" in s:
        t = "Scuola dell'infanzia"
    elif "scuola primaria" in s:
        t = "Scuola Primaria"
    elif ("scuola secondaria di primo " in s) or ("SECONDARIA I" in s):
        t = "Scuola Secondaria di Primo Grado"
    elif ("scuola secondaria di secondo " in s) or ("liceo" in s) or ("istituto tecnico" in s) or ("istituto professionale" in s):
        t = "Scuola Secondaria di Secondo Grado"
    elif ("formazione professionale" in s) or ("formazione professionale" in df.iloc[i,0].lower()):
        t = "Formazione professionale"
    else:
        t = np.nan
    type.append(t)
df['Tipo Istituto'] = type
```

```python
# Map Public and Private school into short version
type = []
for x in df['Tipo Gestione']:
    if "Paritaria" in x:
        type.append("Paritaria")
    else:
        type.append("Statale")
        
df['Tipo Gestione'] = type
```

```python
df.drop_duplicates(['Scuola','Indirizzo'])
```

```python
df = df[['Istituto Principale', 'Scuola', 'Tipo Istituto', 'Tipo Gestione', 'Indirizzo',
       'Comune', 'Telefono', 'Fax', 'Email istituto', 'Email segreteria', 'Sito web', 'Codice MIUR']]
```

## Retrieve schools via OpenStreetMap

```python
import pygeos
import pyrosm
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
trento_download_pbf_url = "https://osmit-estratti.wmcloud.org/dati/poly/province/pbf/022_Trento_poly.osm.pbf"
# download the data
import requests
#request the file
r = requests.get(trento_download_pbf_url, allow_redirects=True)
#save the file
open('../data/trento.pbf', 'wb').write(r.content)
```

```python

# Initialize the OSM object 
osm = pyrosm.OSM("../data/trento.pbf")
```

```python
custom_filter = {'amenity': ['school', 'kindergarten']}
pois = osm.get_pois(custom_filter)
```

```python
pois.explore(marker_type="marker",
             marker_kwds={"radius": "5",
                          "color": "cornflowerblue",
                          'icon': folium.map.Icon(prefix='fa',
                                                  icon='graduation-cap')})
```

```python
osm_schools = gpd.GeoDataFrame(
    pois,
    crs='EPSG:4326',
    geometry=gpd.points_from_xy(pois.lon,
                                pois.lat))
```

```python
import osmnx  as ox
```

```python
def min_dist(point, gpd2):
    gpd2['Dist'] = gpd2.apply(lambda row: point.distance(row.geometry),axis=1)
    closest_index = gpd2.iloc[gpd2['Dist'].argmin()][['geometry','name']]
    return closest_index
```

```python
trentino[['osm_geom','osm_name']] = [min_dist(trentino.iloc[x,-1],osm_schools[~osm_schools['lon'].isnull()])
                     for x in range(len(trentino))]
```
